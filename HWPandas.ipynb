Анализ пизнаков в таблице train_data:
* *row_id* - дублирование индекса, все значения уникальные
* *timestamp* - числовой признак, время в миллисикундах
* *user_id* - числовой признак, id пользователя
* *content_id* - ключ для соединения таблиц
* *content_type_id* - категориальный числовой признак, 2 уникальных значения - 0, если событие было вопросом, заданным пользователю, 1, если событием был просмотр пользователем лекции.
* *task_container_id* - числовой признак
* *user_answer* - категориальный числовой признак, имеет 5 уникальных значений, от -1 до 3, его можно сравнить с признаком *correct_answer* таблицы questions_data
* *answered_correctly* - категориальный числовой признак, имеет 3 уникальных значения, от -1 до 1
* *prior_question_elapsed_time* - числовой признак, среднее время в миллисекундах, которое потребовалось пользователю для ответа на каждый вопрос, также может быть полезен для оценки студента, столбец имеет пропуски
* *prior_question_had_explanation* - признак имеет 2 значения True и False, есть пропуски в столбце
